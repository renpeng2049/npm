## 高斯（正态）分布

正态（或高斯）分布是一种非常常见的连续概率分布。正态分布在统计学中很重要，并且经常在自然科学和社会科学中用于表示分布未知的实值随机变量。具有高斯分布的随机变量被称为正态分布，称为正态偏差。

比如说

$ x \in  R$

```python
# 导入库
import numpy as np
import matplotlib.pyplot as plt
```

```python
mu = 0
sigma = 1 #标准差(Standard Deviation)
data = np.random.normal(mu, sigma, 1000) #生成1000个正态分布随机数
plt.hist(data, bins=30, density=True, alpha=0.5, color='b')
plt.xlabel('value')
plt.ylabel('prob')
plt.title('Normal Distribution')
plt.show()
```

```python
from scipy.stats import norm
pdfx = np.linspace(-4, 4, 100)
pdf = norm.pdf(pdfx, mu, sigma)
plt.plot(pdfx, pdf, color='r', label='概率密度曲线')
plt.show()
```
